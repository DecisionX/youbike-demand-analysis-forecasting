```python
# 3.该部分代码主要是为了得到各个站点在某小时净增加/净减少了多少辆车——即净流量
import pandas as pd

# 讀取乾淨檔案
df = pd.read_csv(
    'data/raw/202410_youbike_clean.csv',
    encoding='utf-8-sig'
)

# ====================================================
# 1. 轉換時間格式
# ====================================================

df['borrow_time'] = pd.to_datetime(df['borrow_time'], errors='coerce')
df['return_time'] = pd.to_datetime(df['return_time'], errors='coerce')
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# ====================================================
# 2. 借車時間特徵
# ====================================================

df['hour'] = df['borrow_time'].dt.hour
df['weekday'] = df['borrow_time'].dt.weekday

# ====================================================
# 3. 新增：還車時間特徵（核心修改）
# ====================================================

df['return_hour'] = df['return_time'].dt.hour
df['return_weekday'] = df['return_time'].dt.weekday

# ====================================================
# 4. 刪除異常數據
# ====================================================

df = df.dropna(subset=['borrow_time', 'return_time'])

# ====================================================
# A. 借車需求
# ====================================================

borrow_hourly = df.groupby(
    ['borrow_station', 'hour', 'weekday', 'date']
).size().reset_index(name='borrow_demand')

# ====================================================
# B. 還車需求（真正基於 return_time）
# ====================================================

return_hourly = df.groupby(
    ['return_station', 'return_hour', 'return_weekday', 'date']
).size().reset_index(name='return_demand')

# ====================================================
# C. 統一欄位名稱
# ====================================================

borrow_hourly = borrow_hourly.rename(columns={
    'borrow_station': 'station'
})

return_hourly = return_hourly.rename(columns={
    'return_station': 'station',
    'return_hour': 'hour',
    'return_weekday': 'weekday'
})

# ====================================================
# D. 合併借車與還車需求
# ====================================================

flow_hourly = pd.merge(
    borrow_hourly,
    return_hourly,
    on=['station', 'hour', 'weekday', 'date'],
    how='outer'
)

# ====================================================
# E. 缺失值補0
# ====================================================

flow_hourly['borrow_demand'] = (
    flow_hourly['borrow_demand'].fillna(0)
)

flow_hourly['return_demand'] = (
    flow_hourly['return_demand'].fillna(0)
)

# ====================================================
# F. 計算淨流量（核心指標）
# ====================================================

flow_hourly['net_flow'] = (
    flow_hourly['borrow_demand']
    - flow_hourly['return_demand']
)

# ====================================================
# G. 保存結果
# ====================================================

borrow_hourly.to_csv(
    'data/processed/borrow_hourly.csv',
    index=False,
    encoding='utf-8-sig'
)

return_hourly.to_csv(
    'data/processed/return_hourly.csv',
    index=False,
    encoding='utf-8-sig'
)

flow_hourly.to_csv(
    'data/processed/flow_hourly.csv',
    index=False,
    encoding='utf-8-sig'
)

# ====================================================
# H. 查看結果
# ====================================================

print("✅ 借車需求、還車需求、淨流量生成完成！")
print(f"處理後共有 {len(flow_hourly):,} 條記錄")
print("\nflow_hourly 前5行：")
print(flow_hourly.head())
```
