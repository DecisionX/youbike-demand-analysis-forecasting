```python
# 6.构建imbalance_score来量化站点潮汐程度的凶险，找出TOP20高风险站点
# 1. 構建 imbalance_score
station_profile['imbalance_score'] = (
    abs(station_profile['morning_net_flow_sum'])
    +
    abs(station_profile['evening_net_flow_sum'])
)

# 2. 風險等級
high_threshold = (
    station_profile['imbalance_score']
    .quantile(0.8)
)

medium_threshold = (
    station_profile['imbalance_score']
    .quantile(0.5)
)

def classify_risk(score):
    if score >= high_threshold:
        return 'High'
    elif score >= medium_threshold:
        return 'Medium'
    else:
        return 'Low'

station_profile['risk_level'] = (
    station_profile['imbalance_score']
    .apply(classify_risk)
)

# 3. 保存風險表
station_profile.to_csv(
    'data/processed/station_risk.csv',
    index=False,
    encoding='utf-8-sig'
)

print("✅ station_risk.csv 已保存")

# 构建Top20 高風險站點
top20_risk = (
    station_profile
    .sort_values(
        by='imbalance_score',
        ascending=False
    )
    .head(20)
)
top20_risk.to_csv(
    'data/processed/top20_risk_stations.csv',
    index=False,
    encoding='utf-8-sig'
)
print("✅ top20_risk_stations.csv 已保存")
print(top20_risk[
    [
        'station',
        'station_type',
        'imbalance_score',
        'risk_level'
    ]
])
```
