```python
import pandas as pd

file_path = 'data/raw/202410_youbike.csv'   # ← 改成你實際的文件名

# 最安全的讀取方式（兼容舊版 pandas）
with open(file_path, 'r', encoding='utf-8-sig', errors='replace') as f:
    df = pd.read_csv(f,
                     header=None,          # 先不認表頭
                     low_memory=False)

# 手動設定正確的列名（根據你上次看到的真實內容）
df.columns = ['borrow_time',      # 借車時間
              'borrow_station',   # 借車站名稱
              'return_time',      # 還車時間
              'return_station',   # 還車站名稱
              'duration',         # 借還時長
              'date']             # 日期

print("✅ Day 1 最終完成！")
print(f"總記錄數：{len(df):,} 條")
print(f"列名已設定：{df.columns.tolist()}")
print("\n前5行數據（現在是正常中文了）：")
print(df.head())

# 保存一份乾淨的原始數據（以後所有分析都用這個檔案）
df.to_csv('data/raw/202410_youbike_clean.csv', index=False, encoding='utf-8-sig')
print("\n✅ 已保存乾淨版本到：data/raw/202410_youbike_clean.csv")
```
