```python
# 1.数据的初步读取以及编码转换工作
import pandas as pd
file_path = 'data/raw/202410_youbike.csv'   # ← 改成你實際的文件名
# 最安全的讀取方式（兼容舊版 pandas）
with open(file_path, 'r', encoding='utf-8-sig', errors='replace') as f:
    df = pd.read_csv(f,
                     header=None,          # 先不認表頭
                     low_memory=False)
# 手動設定正確的列名（根據你上次看到的真實內容）
df.columns = ['borrow_time',      # 借車時間
              'borrow_station',   # 借車站名稱
              'return_time',      # 還車時間
              'return_station',   # 還車站名稱
              'duration',         # 借還時長
              'date']             # 日期
print("✅ Day 1 最終完成！")
print(f"總記錄數：{len(df):,} 條")
print(f"列名已設定：{df.columns.tolist()}")
print("\n前5行數據（現在是正常中文了）：")
print(df.head())

# 保存一份乾淨的原始數據（以後所有分析都用這個檔案）
df.to_csv('data/raw/202410_youbike_clean.csv', index=False, encoding='utf-8-sig')
print("\n✅ 已保存乾淨版本到：data/raw/202410_youbike_clean.csv")
```

```python
# 2.输出hourly_demand.csv:各站点在不同时间某个小时的借车量
import pandas as pd
# 讀取我們剛才保存的乾淨檔案
df = pd.read_csv('data/raw/202410_youbike_clean.csv', encoding='utf-8-sig')

# 1. 轉換時間格式
df['borrow_time'] = pd.to_datetime(df['borrow_time'], errors='coerce')
df['return_time'] = pd.to_datetime(df['return_time'], errors='coerce')
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# 2. 提取小時和星期幾
df['hour'] = df['borrow_time'].dt.hour
df['weekday'] = df['borrow_time'].dt.weekday   # 0=周一 ... 6=周日

# 3. 刪除異常數據（即borrow_time=return_time的异常数据）
df = df.dropna(subset=['borrow_time', 'return_time'])

# 4. 聚合：每個站點每小時的借車需求量（這就是我們後面分析的核心數據）
demand_hourly = df.groupby(['borrow_station', 'hour', 'weekday', 'date']).size().reset_index(name='demand')

# 5. 保存處理好的數據
demand_hourly.to_csv('data/processed/demand_hourly.csv', index=False, encoding='utf-8-sig')

print("✅ Day 2 數據清洗完成！")
print(f"處理後共有 {len(demand_hourly):,} 條記錄")
print("\n前5行結果：")
print(demand_hourly.head())
```

1.pd.to_datetime将原始为字符串的时间转化成为了datetime，以便于后续对星期几和小时的提取。
2.之所以要提取出hour和weekday是因为共享单车的需求最主要受到两个影响因素，一个是几点钟（例如高峰期VS半夜），另一个就是星期几（工作日VS周末）。
3.df.dropna用于删除异常值，因为原始数据里有借车时间和换车时间相同的这类无效数据，留着的话会对后续的统计结果产生污染。。
4.而之所以要采用groupby是为了得到整体的一个需求规律，以便于得到需要的每小时需求曲线以及预测模型，从而给出未来的需求建议。如果不groupby那只能看到某一笔交易，而无法得到整体的需求估计。
5.最终保存为demand_hourly.csv是以便于未来所有的画图，预测以及报告都只用这个小文件，不再使用那个五百多万行的原始数据，代码的运行速度也会快上很多。

总的来说这第二段代码的主要作用就是要得到“每个站点在某个小时借了多少辆车”以便于得到各站点单车的历史需求量来进行对未来提前的储备量进做准备。
总的来说，就是将“散乱的每一笔借车记录”加工为“结构化的每小时需求表”，这也是后续进行商业预测的地基。

```python
# 3.该部分代码主要是为了得到各个站点在某小时净增加/净减少了多少辆车——即净流量
import pandas as pd

# 讀取乾淨檔案
df = pd.read_csv(
    'data/raw/202410_youbike_clean.csv',
    encoding='utf-8-sig'
)

# ====================================================
# 1. 轉換時間格式
# ====================================================

df['borrow_time'] = pd.to_datetime(df['borrow_time'], errors='coerce')
df['return_time'] = pd.to_datetime(df['return_time'], errors='coerce')
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# ====================================================
# 2. 借車時間特徵
# ====================================================

df['hour'] = df['borrow_time'].dt.hour
df['weekday'] = df['borrow_time'].dt.weekday

# ====================================================
# 3. 新增：還車時間特徵（核心修改）
# ====================================================

df['return_hour'] = df['return_time'].dt.hour
df['return_weekday'] = df['return_time'].dt.weekday

# ====================================================
# 4. 刪除異常數據
# ====================================================

df = df.dropna(subset=['borrow_time', 'return_time'])

# ====================================================
# A. 借車需求
# ====================================================

borrow_hourly = df.groupby(
    ['borrow_station', 'hour', 'weekday', 'date']
).size().reset_index(name='borrow_demand')

# ====================================================
# B. 還車需求（真正基於 return_time）
# ====================================================

return_hourly = df.groupby(
    ['return_station', 'return_hour', 'return_weekday', 'date']
).size().reset_index(name='return_demand')

# ====================================================
# C. 統一欄位名稱
# ====================================================

borrow_hourly = borrow_hourly.rename(columns={
    'borrow_station': 'station'
})

return_hourly = return_hourly.rename(columns={
    'return_station': 'station',
    'return_hour': 'hour',
    'return_weekday': 'weekday'
})

# ====================================================
# D. 合併借車與還車需求
# ====================================================

flow_hourly = pd.merge(
    borrow_hourly,
    return_hourly,
    on=['station', 'hour', 'weekday', 'date'],
    how='outer'
)

# ====================================================
# E. 缺失值補0
# ====================================================

flow_hourly['borrow_demand'] = (
    flow_hourly['borrow_demand'].fillna(0)
)

flow_hourly['return_demand'] = (
    flow_hourly['return_demand'].fillna(0)
)

# ====================================================
# F. 計算淨流量（核心指標）
# ====================================================

flow_hourly['net_flow'] = (
    flow_hourly['borrow_demand']
    - flow_hourly['return_demand']
)

# ====================================================
# G. 保存結果
# ====================================================

borrow_hourly.to_csv(
    'data/processed/borrow_hourly.csv',
    index=False,
    encoding='utf-8-sig'
)

return_hourly.to_csv(
    'data/processed/return_hourly.csv',
    index=False,
    encoding='utf-8-sig'
)

flow_hourly.to_csv(
    'data/processed/flow_hourly.csv',
    index=False,
    encoding='utf-8-sig'
)

# ====================================================
# H. 查看結果
# ====================================================

print("✅ 借車需求、還車需求、淨流量生成完成！")
print(f"處理後共有 {len(flow_hourly):,} 條記錄")
print("\nflow_hourly 前5行：")
print(flow_hourly.head())
```

```python
# 5.对所有站点进行分层，共有四个类型：通勤型站点，潮汐流入型站点，潮汐流出型站点，稳定休闲型站点。
# YouBike 站點画像與分層分析
# ====================================================
import pandas as pd
import numpy as np
# 1. 讀取 flow_hourly
df = pd.read_csv(
    'data/processed/flow_hourly.csv',
    encoding='utf-8-sig'
)

# 2. 定義高峰期
morning_peak = [7, 8, 9]
evening_peak = [16, 17, 18, 19]

# 3. 建立站點画像
station_profile = []
stations = df['station'].unique()
for station in stations:
    temp = df[df['station'] == station]

    # 全天總借車量
    total_borrow = temp['borrow_demand'].sum()

    # 早晚高峰期的总借車量
    morning_borrow = temp[
        temp['hour'].isin(morning_peak)
    ]['borrow_demand'].sum()

    evening_borrow = temp[
        temp['hour'].isin(evening_peak)
    ]['borrow_demand'].sum()


    # 周末需求
    weekend_borrow = temp[
        temp['weekday'].isin([5, 6])
    ]['borrow_demand'].sum()

    # 早晚高峰期的平均每小時淨流量
    morning_net_flow_mean = temp[
        temp['hour'].isin(morning_peak)
    ]['net_flow'].mean()

    evening_net_flow_mean = temp[
        temp['hour'].isin(evening_peak)
    ]['net_flow'].mean()

    # 早晚高峰總淨流量
    morning_net_flow_sum = temp[
        temp['hour'].isin(morning_peak)
    ]['net_flow'].sum()

    evening_net_flow_sum = temp[
        temp['hour'].isin(evening_peak)
    ]['net_flow'].sum()

    # 波動率
    volatility = temp['borrow_demand'].std()

    # 高峰占比
    peak_ratio = ( # 早晚高峰期的借车量在一整天借车量中的占比
        (morning_borrow + evening_borrow)
        / total_borrow
        if total_borrow != 0 else 0
    )

    # 周末占比
    weekend_ratio = ( # 周末借车量在一整周借车量中的占比
        weekend_borrow / total_borrow
        if total_borrow != 0 else 0
    )
    
    # 保存站點画像
    station_profile.append({
        'station': station,
        'total_borrow': total_borrow,
        'peak_ratio': peak_ratio, # 早晚高峰期的借车量在一整天借车量中的占比
        'weekend_ratio': weekend_ratio, # 周末借车量在一整周借车量中的占比
        'morning_net_flow_mean': morning_net_flow_mean,
        'evening_net_flow_mean': evening_net_flow_mean,
        'morning_net_flow_sum': morning_net_flow_sum,
        'evening_net_flow_sum': evening_net_flow_sum,
        'volatility': volatility
    })
# 4. 轉 DataFrame
station_profile = pd.DataFrame(station_profile)

# 5. 規則分類
upper_morning_threshold = (
    station_profile['morning_net_flow_mean']
    .quantile(0.8)
)
lower_morning_threshold = (
    station_profile['morning_net_flow_mean']
    .quantile(0.2)
)
upper_evening_threshold = (
    station_profile['evening_net_flow_mean']
    .quantile(0.8)
)
lower_evening_threshold = (
    station_profile['evening_net_flow_mean']
    .quantile(0.2)
)
print("早高峰上阈值：", upper_morning_threshold)
print("早高峰下阈值：", lower_morning_threshold)
print("晚高峰上阈值：", upper_evening_threshold)
print("晚高峰下阈值：", lower_evening_threshold)

peak_threshold = (
    station_profile['peak_ratio']
    .quantile(0.7)
)

def classify_station(row):

    # 潮汐流出型（住宅區）
    if (
        row['morning_net_flow_mean'] > upper_morning_threshold
        and
        row['evening_net_flow_mean'] < lower_evening_threshold
    ):
        return '潮汐流出型'

    # 潮汐流入型（CBD）
    elif (
        row['morning_net_flow_mean'] < lower_morning_threshold
        and
        row['evening_net_flow_mean'] > upper_evening_threshold
    ):
        return '潮汐流入型'

    # 通勤型
    elif row['peak_ratio'] > peak_threshold:
        return '通勤型'

    # 穩定休閒型
    else:
        return '穩定休閒型'


station_profile['station_type'] = (
    station_profile.apply(classify_station, axis=1)
)

# ====================================================
# 6. 合併回原始小時表
# ====================================================

station_type_hourly = pd.merge(
    df,
    station_profile[['station', 'station_type']],
    on='station',
    how='left'
)

# ====================================================
# 7. 導出 CSV
# ====================================================

station_profile.to_csv(
    'data/processed/station_profile.csv',
    index=False,
    encoding='utf-8-sig'
)

station_type_hourly.to_csv(
    'data/processed/station_type_hourly.csv',
    index=False,
    encoding='utf-8-sig'
)

print("✅ station_profile.csv 已保存")
print("✅ station_type_hourly.csv 已保存")

# ====================================================
# 8. 查看分類結果
# ====================================================

print("\n站點類型分布：")
print(
    station_profile['station_type'].value_counts()
)

print("\nstation_profile 前5行：")
print(station_profile.head())
```

```python
# 6.构建imbalance_score来量化站点潮汐程度的凶险，找出TOP20高风险站点
# 1. 構建 imbalance_score
station_profile['imbalance_score'] = (
    abs(station_profile['morning_net_flow_sum'])
    +
    abs(station_profile['evening_net_flow_sum'])
)

# 2. 風險等級
high_threshold = (
    station_profile['imbalance_score']
    .quantile(0.8)
)

medium_threshold = (
    station_profile['imbalance_score']
    .quantile(0.5)
)

def classify_risk(score):
    if score >= high_threshold:
        return 'High'
    elif score >= medium_threshold:
        return 'Medium'
    else:
        return 'Low'

station_profile['risk_level'] = (
    station_profile['imbalance_score']
    .apply(classify_risk)
)

# 3. 保存風險表
station_profile.to_csv(
    'data/processed/station_risk.csv',
    index=False,
    encoding='utf-8-sig'
)

print("✅ station_risk.csv 已保存")

# 构建Top20 高風險站點
top20_risk = (
    station_profile
    .sort_values(
        by='imbalance_score',
        ascending=False
    )
    .head(20)
)
top20_risk.to_csv(
    'data/processed/top20_risk_stations.csv',
    index=False,
    encoding='utf-8-sig'
)
print("✅ top20_risk_stations.csv 已保存")
print(top20_risk[
    [
        'station',
        'station_type',
        'imbalance_score',
        'risk_level'
    ]
])
```

```python
# 7.調度建議系統
def dispatch_strategy(row):
    # （1）高風險站點
    if row['risk_level'] == 'High':
        # 潮汐流出型
        if row['station_type'] == '潮汐流出型':
            return '早高峰前提前補車'

        # 潮汐流入型
        elif row['station_type'] == '潮汐流入型':
            return '高峰前提前清空車位'

        # 通勤型
        elif row['station_type'] == '通勤型':
            return '雙高峰動態調度'

        # 穩定休閒型
        else:
            return '周末增加車輛配置'

    # （2）中風險站點
    elif row['risk_level'] == 'Medium':
        return '定時監控與小規模調度'

    # （3）低風險站點
    else:
        return '常規運營監控'

# 生成調度建議
station_profile['dispatch_strategy'] = (

    station_profile
    .apply(dispatch_strategy, axis=1)
)
station_profile.to_csv(
    'data/processed/dispatch_strategy.csv',
    index=False,
    encoding='utf-8-sig'
)
print("✅ dispatch_strategy.csv 已保存")
print(
    station_profile[
        [
            'station',
            'station_type',
            'risk_level',
            'dispatch_strategy'
        ]
    ].head()
)
```

```python
# 8.根据历史数据，通过prophet模型来预测未来七天单车的
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# ====================== 解决中文显示问题 ======================
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False
# ============================================================

# 读取清理后的数据
df = pd.read_csv('data/processed/demand_hourly.csv', encoding='utf-8-sig')

print("✅ Day 5 预测模型开始运行...")

# Prophet需要特定格式：ds（日期时间） + y（预测目标）
daily_hourly = df.groupby(['date', 'hour'])['demand'].sum().reset_index()
daily_hourly['ds'] = pd.to_datetime(daily_hourly['date']) + pd.to_timedelta(daily_hourly['hour'], unit='h')
daily_hourly = daily_hourly.rename(columns={'demand': 'y'})

prophet_df = daily_hourly[['ds', 'y']]

# 训练模型
model = Prophet(daily_seasonality=True, weekly_seasonality=True)
model.fit(prophet_df)

# 预测未来7天（168小时）——这里改成小写 'h'
future = model.make_future_dataframe(periods=7*24, freq='h')
forecast = model.predict(future)

# 显示预测结果
print("\n✅ 预测完成！未来7天部分预测结果：")
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(15))

# 保存预测图
fig = model.plot(forecast)
plt.title('YB单车未来7天每小时需求预测（含置信区间）')
plt.xlabel('日期')
plt.ylabel('预测借车需求量')
plt.savefig('reports/figures/forecast_7days.png', dpi=300, bbox_inches='tight')
plt.show()

# 保存预测数据（后面写报告会用到）
forecast.to_csv('data/processed/forecast_7days.csv', index=False, encoding='utf-8-sig')
print("\n✅ 预测图和数据已保存！")
```
