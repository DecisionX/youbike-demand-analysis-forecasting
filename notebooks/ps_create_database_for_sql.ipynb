```python
import pandas as pd
import sqlite3

# ====================================================
# 1. 建立数据库连接
# ====================================================

conn = sqlite3.connect(
    'data/youbike_warehouse.db'
)

print("✅ 数据库连接成功")

# ====================================================
# 2. 导入原始骑行数据
# ====================================================

print("正在读取原始 CSV 数据...")

df = pd.read_csv(
    'data/raw/202410_youbike_clean.csv',
    low_memory=False
)

print("正在导入 clean_youbike_log...")

df.to_sql(
    'clean_youbike_log',
    conn,
    if_exists='replace',
    index=False
)

print("✅ clean_youbike_log 导入成功")

# ====================================================
# 3. 导入 flow_hourly
# ====================================================

flow_hourly = pd.read_csv(
    'data/processed/flow_hourly.csv',
    encoding='utf-8-sig'
)

flow_hourly.to_sql(
    'flow_hourly',
    conn,
    if_exists='replace',
    index=False
)

print("✅ flow_hourly 导入成功")

# ====================================================
# 4. 导入 station_profile
# ====================================================

station_profile = pd.read_csv(
    'data/processed/station_profile.csv',
    encoding='utf-8-sig'
)

station_profile.to_sql(
    'station_profile',
    conn,
    if_exists='replace',
    index=False
)

print("✅ station_profile 导入成功")

# ====================================================
# 5. 导入 station_risk
# ====================================================

station_risk = pd.read_csv(
    'data/processed/station_risk.csv',
    encoding='utf-8-sig'
)

station_risk.to_sql(
    'station_risk',
    conn,
    if_exists='replace',
    index=False
)

print("✅ station_risk 导入成功")

# ====================================================
# 6. 导入 dispatch_strategy
# ====================================================

dispatch_strategy = pd.read_csv(
    'data/processed/dispatch_strategy.csv',
    encoding='utf-8-sig'
)

dispatch_strategy.to_sql(
    'dispatch_strategy',
    conn,
    if_exists='replace',
    index=False
)

print("✅ dispatch_strategy 导入成功")

# ====================================================
# 7. 关闭数据库
# ====================================================

conn.close()

print("✅ 所有数据表导入完成！")
```
