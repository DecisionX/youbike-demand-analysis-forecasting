```python
# 5.对所有站点进行分层，共有四个类型：通勤型站点，潮汐流入型站点，潮汐流出型站点，稳定休闲型站点。
# YouBike 站點画像與分層分析
# ====================================================
import pandas as pd
import numpy as np
# 1. 讀取 flow_hourly
df = pd.read_csv(
    'data/processed/flow_hourly.csv',
    encoding='utf-8-sig'
)

# 2. 定義高峰期
morning_peak = [7, 8, 9]
evening_peak = [16, 17, 18, 19]

# 3. 建立站點画像
station_profile = []
stations = df['station'].unique()
for station in stations:
    temp = df[df['station'] == station]

    # 全天總借車量
    total_borrow = temp['borrow_demand'].sum()

    # 早晚高峰期的总借車量
    morning_borrow = temp[
        temp['hour'].isin(morning_peak)
    ]['borrow_demand'].sum()

    evening_borrow = temp[
        temp['hour'].isin(evening_peak)
    ]['borrow_demand'].sum()


    # 周末需求
    weekend_borrow = temp[
        temp['weekday'].isin([5, 6])
    ]['borrow_demand'].sum()

    # 早晚高峰期的平均每小時淨流量
    morning_net_flow_mean = temp[
        temp['hour'].isin(morning_peak)
    ]['net_flow'].mean()

    evening_net_flow_mean = temp[
        temp['hour'].isin(evening_peak)
    ]['net_flow'].mean()

    # 早晚高峰總淨流量
    morning_net_flow_sum = temp[
        temp['hour'].isin(morning_peak)
    ]['net_flow'].sum()

    evening_net_flow_sum = temp[
        temp['hour'].isin(evening_peak)
    ]['net_flow'].sum()

    # 波動率
    volatility = temp['borrow_demand'].std()

    # 高峰占比
    peak_ratio = ( # 早晚高峰期的借车量在一整天借车量中的占比
        (morning_borrow + evening_borrow)
        / total_borrow
        if total_borrow != 0 else 0
    )

    # 周末占比
    weekend_ratio = ( # 周末借车量在一整周借车量中的占比
        weekend_borrow / total_borrow
        if total_borrow != 0 else 0
    )

    # 保存站點画像
    station_profile.append({
        'station': station,
        'total_borrow': total_borrow,
        'peak_ratio': peak_ratio, # 早晚高峰期的借车量在一整天借车量中的占比
        'weekend_ratio': weekend_ratio, # 周末借车量在一整周借车量中的占比
        'morning_net_flow_mean': morning_net_flow_mean,
        'evening_net_flow_mean': evening_net_flow_mean,
        'morning_net_flow_sum': morning_net_flow_sum,
        'evening_net_flow_sum': evening_net_flow_sum,
        'volatility': volatility
    })
# 4. 轉 DataFrame
station_profile = pd.DataFrame(station_profile)

# 5. 規則分類
upper_morning_threshold = (
    station_profile['morning_net_flow_mean']
    .quantile(0.8)
)
lower_morning_threshold = (
    station_profile['morning_net_flow_mean']
    .quantile(0.2)
)
upper_evening_threshold = (
    station_profile['evening_net_flow_mean']
    .quantile(0.8)
)
lower_evening_threshold = (
    station_profile['evening_net_flow_mean']
    .quantile(0.2)
)
print("早高峰上阈值：", upper_morning_threshold)
print("早高峰下阈值：", lower_morning_threshold)
print("晚高峰上阈值：", upper_evening_threshold)
print("晚高峰下阈值：", lower_evening_threshold)

peak_threshold = (
    station_profile['peak_ratio']
    .quantile(0.7)
)

def classify_station(row):

    # 潮汐流出型（住宅區）
    if (
        row['morning_net_flow_mean'] > upper_morning_threshold
        and
        row['evening_net_flow_mean'] < lower_evening_threshold
    ):
        return '潮汐流出型'

    # 潮汐流入型（CBD）
    elif (
        row['morning_net_flow_mean'] < lower_morning_threshold
        and
        row['evening_net_flow_mean'] > upper_evening_threshold
    ):
        return '潮汐流入型'

    # 通勤型
    elif row['peak_ratio'] > peak_threshold:
        return '通勤型'

    # 穩定休閒型
    else:
        return '穩定休閒型'


station_profile['station_type'] = (
    station_profile.apply(classify_station, axis=1)
)

# ====================================================
# 6. 合併回原始小時表
# ====================================================

station_type_hourly = pd.merge(
    df,
    station_profile[['station', 'station_type']],
    on='station',
    how='left'
)

# ====================================================
# 7. 導出 CSV
# ====================================================

station_profile.to_csv(
    'data/processed/station_profile.csv',
    index=False,
    encoding='utf-8-sig'
)

station_type_hourly.to_csv(
    'data/processed/station_type_hourly.csv',
    index=False,
    encoding='utf-8-sig'
)

print("✅ station_profile.csv 已保存")
print("✅ station_type_hourly.csv 已保存")

# ====================================================
# 8. 查看分類結果
# ====================================================

print("\n站點類型分布：")
print(
    station_profile['station_type'].value_counts()
)

print("\nstation_profile 前5行：")
print(station_profile.head())
```
