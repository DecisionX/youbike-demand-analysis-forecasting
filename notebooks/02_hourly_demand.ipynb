```python
# 2.输出hourly_demand.csv:各站点在不同时间某个小时的借车量
import pandas as pd
# 讀取我們剛才保存的乾淨檔案
df = pd.read_csv('data/raw/202410_youbike_clean.csv', encoding='utf-8-sig')

# 1. 轉換時間格式
df['borrow_time'] = pd.to_datetime(df['borrow_time'], errors='coerce')
df['return_time'] = pd.to_datetime(df['return_time'], errors='coerce')
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# 2. 提取小時和星期幾
df['hour'] = df['borrow_time'].dt.hour
df['weekday'] = df['borrow_time'].dt.weekday   # 0=周一 ... 6=周日

# 3. 刪除異常數據（即borrow_time=return_time的异常数据）
df = df.dropna(subset=['borrow_time', 'return_time'])

# 4. 聚合：每個站點每小時的借車需求量（這就是我們後面分析的核心數據）
demand_hourly = df.groupby(['borrow_station', 'hour', 'weekday', 'date']).size().reset_index(name='demand')

# 5. 保存處理好的數據
demand_hourly.to_csv('data/processed/demand_hourly.csv', index=False, encoding='utf-8-sig')

print("✅ Day 2 數據清洗完成！")
print(f"處理後共有 {len(demand_hourly):,} 條記錄")
print("\n前5行結果：")
print(demand_hourly.head())
```
