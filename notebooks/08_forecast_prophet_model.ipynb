```python
# 8.根据历史数据，通过prophet模型来预测未来七天单车的
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# ====================== 解决中文显示问题 ======================
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False
# ============================================================

# 读取清理后的数据
df = pd.read_csv('data/processed/demand_hourly.csv', encoding='utf-8-sig')

print("✅ Day 5 预测模型开始运行...")

# Prophet需要特定格式：ds（日期时间） + y（预测目标）
daily_hourly = df.groupby(['date', 'hour'])['demand'].sum().reset_index()
daily_hourly['ds'] = pd.to_datetime(daily_hourly['date']) + pd.to_timedelta(daily_hourly['hour'], unit='h')
daily_hourly = daily_hourly.rename(columns={'demand': 'y'})

prophet_df = daily_hourly[['ds', 'y']]

# 训练模型
model = Prophet(daily_seasonality=True, weekly_seasonality=True)
model.fit(prophet_df)

# 预测未来7天（168小时）——这里改成小写 'h'
future = model.make_future_dataframe(periods=7*24, freq='h')
forecast = model.predict(future)

# 显示预测结果
print("\n✅ 预测完成！未来7天部分预测结果：")
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(15))

# 保存预测图
fig = model.plot(forecast)
plt.title('YB单车未来7天每小时需求预测（含置信区间）')
plt.xlabel('日期')
plt.ylabel('预测借车需求量')
plt.savefig('reports/figures/forecast_7days.png', dpi=300, bbox_inches='tight')
plt.show()

# 保存预测数据（后面写报告会用到）
forecast.to_csv('data/processed/forecast_7days.csv', index=False, encoding='utf-8-sig')
print("\n✅ 预测图和数据已保存！")
```
