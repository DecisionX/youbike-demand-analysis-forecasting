```python
# 7.調度建議系統
def dispatch_strategy(row):
    # （1）高風險站點
    if row['risk_level'] == 'High':
        # 潮汐流出型
        if row['station_type'] == '潮汐流出型':
            return '早高峰前提前補車'

        # 潮汐流入型
        elif row['station_type'] == '潮汐流入型':
            return '高峰前提前清空車位'

        # 通勤型
        elif row['station_type'] == '通勤型':
            return '雙高峰動態調度'

        # 穩定休閒型
        else:
            return '周末增加車輛配置'

    # （2）中風險站點
    elif row['risk_level'] == 'Medium':
        return '定時監控與小規模調度'

    # （3）低風險站點
    else:
        return '常規運營監控'

# 生成調度建議
station_profile['dispatch_strategy'] = (

    station_profile
    .apply(dispatch_strategy, axis=1)
)
station_profile.to_csv(
    'data/processed/dispatch_strategy.csv',
    index=False,
    encoding='utf-8-sig'
)
print("✅ dispatch_strategy.csv 已保存")
print(
    station_profile[
        [
            'station',
            'station_type',
            'risk_level',
            'dispatch_strategy'
        ]
    ].head()
)
```
