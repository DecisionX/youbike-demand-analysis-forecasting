```python
# 4.探究早晚高峰期，工作日vs周末，不同站点这三大因素对单车需求量的影响程度
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# ====================== 解决中文显示问题 ======================
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']   # Windows常用中文字体
plt.rcParams['axes.unicode_minus'] = False               # 解决负号显示问题
# ============================================================

print("✅ 中文字体已设置，后续不再出现警告")
# 读取清理后的干净数据
df = pd.read_csv('data/processed/demand_hourly.csv', encoding='utf-8-sig')

print("✅ Day 3 开始！数据载入成功")

# ==================== 图1：每小时平均需求（最重要） ====================
hourly_avg = df.groupby('hour')['demand'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=hourly_avg, x='hour', y='demand', marker='o')
plt.title('每小时平均借车需求量（台北YouBike 2024年10月）')
plt.xlabel('小时 (0-23)')
plt.ylabel('平均需求量')
plt.grid(True)
plt.savefig('reports/figures/hourly_demand.png', dpi=300, bbox_inches='tight')
plt.show()

# ==================== 图2：星期几需求 ====================
weekday_avg = df.groupby('weekday')['demand'].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=weekday_avg, x='weekday', y='demand')
plt.title('不同星期平均借车需求')
plt.xlabel('星期 (0=周一, 6=周日)')
plt.ylabel('平均需求量')
plt.savefig('reports/figures/weekday_demand.png', dpi=300, bbox_inches='tight')
plt.show()

# ==================== 图3：Top 10 最热门站点 ====================
top10 = df.groupby('borrow_station')['demand'].sum().nlargest(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top10.values, y=top10.index)
plt.title('Top 10 最热门借车站点（10月总需求）')
plt.xlabel('总借车次数')
plt.ylabel('站点名称')
plt.tight_layout()
plt.savefig('reports/figures/top10_stations.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✅ 3张图已生成并保存到 reports/ 文件夹！")
print("\nTop 10 最热门站点：")
print(top10)
```
